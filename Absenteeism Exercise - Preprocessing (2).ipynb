```python
# import data

import pandas as pd
raw_csv_data = pd.read_csv("Absenteeism_data.csv")
raw_csv_data
```

```python
# setup df view preferences

# pd.options.display.max_columns = None
# pd.options.display.max_rows = None
```

```python
#copy df

df = raw_csv_data.copy()
```

```python
# eyeball  df

df.info()
```

```python
#start preprocessing
# column #1 ID

df = df.drop(['ID'], axis = 1) #column drop
```

```python
# column #2 Reason for Absence
#extract the column reason for absence
df['Reason for Absence'].min()
```

```python
df['Reason for Absence'].max()
```

```python
len(df['Reason for Absence'].unique())
```

```python
sorted(df['Reason for Absence'].unique())
```

```python
# transform to dummies for analysis purposes
#.get_dummies()
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns['check'].unique()
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first =True)
reason_columns
```

```python
# grouping the dummies to create more relevant prediction preparations
# get column names
df.columns.values
```

```python
reason_columns.columns.values
```

```python
#drop the original column

df = df.drop(['Reason for Absence'], axis = 1)
df
```

```python
reason_type_1 = reason_columns.loc[:, '1':'14'].max(axis=1)
reason_type_2 = reason_columns.loc[:, '15':'17'].max(axis=1)
reason_type_3 = reason_columns.loc[:, '18':'21'].max(axis=1)
reason_type_4 = reason_columns.loc[:, '22':].max(axis=1)
```

```python
#concat
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df
```

```python
df.columns.values
```

```python
#update column names
df.columns = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.head()
```

```python
#reorder column names
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
df.head()
```

```python
#create a checkpoint1: a copy of the current state of the df

df_reason_mod = df.copy()
df_reason_mod
```

```python
#work on the next column: column#3 date

type(df_reason_mod['Date'][0])
```

```python
#using to_datetime, extract relevant date information. note: format

df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod['Date']
```

```python
#extract the month value
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0]
```

```python
#get the table size/ the number of rows
df_reason_mod.shape
```

```python
#create a column with month data
list_months = []

for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

```python
#extract the day value
df_reason_mod['Date'][699].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday) 
```

```python
#create checkpoint2
df_reason_date_mod = df_reason_mod.copy()
df_reason_date_mod.head()
```

```python
# column#4: education

df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts()
```

```python
#grouping 1, 2-4
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```
